## Features based on correlation

```python
quantile_f4 = train.f4.quantile(q=0.2)

def create_special(row):
    return int((row['f4'] < quantile_f4))

train.loc[:, 'special'] = train.apply(create_special, axis=1)
```

```python
pd.crosstab(train.special, train.target)
```

## K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
X = train[train.columns[:-3]]
y = train.target
```

```python
kmeans = KMeans(n_clusters=2, n_jobs=-1, n_init=10)

%time kmeans.fit(X)
```

```python
train.loc[:, 'cluster_labels'] = kmeans.labels_
```

```python
predict = kmeans.predict(X)
```

```python
pd.crosstab(predict, train.target)
```

```python
kmeans.transform(X)
```
